문제  
카지노에서 제일 인기 있는 게임 블랙잭의 규칙은 상당히 쉽다. 카드의 합이 21을 넘지 않는 한도 내에서, 카드의 합을 최대한 크게 만드는 게임이다. 블랙잭은 카지노마다 다양한 규정이 있다.  

한국 최고의 블랙잭 고수 김정인은 새로운 블랙잭 규칙을 만들어 상근, 창영이와 게임하려고 한다.  

김정인 버전의 블랙잭에서 각 카드에는 양의 정수가 쓰여 있다. 그 다음, 딜러는 N장의 카드를 모두 숫자가 보이도록 바닥에 놓는다. 그런 후에 딜러는 숫자 M을 크게 외친다.  

이제 플레이어는 제한된 시간 안에 N장의 카드 중에서 3장의 카드를 골라야 한다. 블랙잭 변형 게임이기 때문에, 플레이어가 고른 카드의 합은 M을 넘지 않으면서 M과 최대한 가깝게 만들어야 한다.  

N장의 카드에 써져 있는 숫자가 주어졌을 때, M을 넘지 않으면서 M에 최대한 가까운 카드 3장의 합을 구해 출력하시오.  

입력  
첫째 줄에 카드의 개수 N(3 ≤ N ≤ 100)과 M(10 ≤ M ≤ 300,000)이 주어진다. 둘째 줄에는 카드에 쓰여 있는 수가 주어지며, 이 값은 100,000을 넘지 않는 양의 정수이다.  

합이 M을 넘지 않는 카드 3장을 찾을 수 있는 경우만 입력으로 주어진다.  

출력   
첫째 줄에 M을 넘지 않으면서 M에 최대한 가까운 카드 3장의 합을 출력한다.  

```python
# 실패 : 메모리 초과
def select_card(cards,M):
    sum_list = cards
    card1 = cards[0]
    for card2 in cards[1:]:
        sum2 = card1 + card2

        if sum2 == M: # 두개만으로 M
            return M

        elif sum2 < M: # 두개로 M 안될시 card3 추가
            for card3 in cards[2:]:
                sum3 = card1 + card2 + card3 
                if sum3 == M:
                    return M
                elif sum3 < M:
                    sum_list.append(sum3)

                else: 
                    sum_list.append(sum2)
                    break
        else:
            break
    return max(sum_list)

N,M = map(int,input().split())
cards = list(map(int,input().split()))
cards.sort()
select_card(cards, M)

```

```python
# 틀렸습니다.

def select_card(cards,M):
    sum_list = []
    for card1 in cards:
        for card2 in cards:
            for card3 in cards:
                if card1 == card2 or card1 == card3 or card2 == card3:
                    pass
                else:
                    cards_sum = card1 + card2 + card3
                    if cards_sum<M:
                        sum_list.append(cards_sum)
                    elif cards_sum == M:
                        return cards_sum
                    else:
                        break
    return max(sum_list)

N,M = map(int,input().split())
cards = list(map(int,input().split()))
cards.sort()
select_card(cards, M)
```

```python
# 다른사람 풀이
# 완전탐색  콤비내이션을 range수정을 통해 구현
N,M = map(int,input().split())
cards = list(map(int,input().split()))
result = 0

for i in range(N):# 길이 만큼
    for j in range(i+1,N): # i+1 ~ N
        for k in range(j+1, N): # j+1 ~ N
            if cards[i]+cards[j]+cards[k] > M:
                continue
            else:
                result = max(result, cards[i]+cards[j]+cards[k])    

print(result) 
```

```python
# 다른사람 풀이
# 콤비내이션 라이브러리 사용

from itertools import combinations

N,M = map(int,input().split())
cards_list = list(map(int,input().split()))
result = 0

for cards in combinations(cards_list,3): # combinations(iter, num) -> iter에서 num개 뽑은 콤비내이션 iterator객체 생성
    if sum(cards) <= M:
        result = max(result, sum(cards))

print(result)
```

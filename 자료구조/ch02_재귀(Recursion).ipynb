```python
#제귀함수 기초 (탈출조건!)
def recursive_func_basic(n):
    print(n)
    if n==1:   # 탈출조건!
        return
    recursive_func_basic(n-1)
    
recursive_func_basic(5)
```

```python
#재귀함수 피보나치수열
def fibonachi(n):
    if n==1:
        return 0
    elif n==2:
        return 1
    else:
        return fibonachi(n-1)+fibonachi(n-2)

fibonachi(3)
```

```python
#재귀함수 팩토리얼
def factorial(n):
    if n == 0:
        return 1
    else:
        return n*factorial(n-1)

factorial(3)
```

```python
#재귀함수 바이너리 서치 (성능은 떨어지지만 공부용으루)
def Bsearch(lst,first,last,target):
    mid=(first+last)//2
    if first<=last:
        if target==lst[mid]:
            return mid
        else:
            if target<lst[mid]:
                return Bsearch(lst,first,mid-1,target)
            elif target>lst[mid]:
                return Bsearch(lst,mid+1,last,target)

    else:
        print("없서욧!")
        
        
lst=list(range(0,20,2))
target=7
first=0
last=len(lst)-1

print(lst[0:])
Bsearch(lst,first,last,target)
```

###### 재귀적인 패턴을 찾는것이 우선!
- 작은것 두개로 큰것하나 해결 가능 패턴
- 4개옮기는 문제 -> 3개옮기는 문제 -> 2개옮기는 문제 -> 1개옮기는 문제 
- n개옮기는 문제 -> n-1개옮기는 문제  

###### 일반화  
  
목적 : 큰원반 n개를 A->C로
 1. 작은원반 n-1개를 A->B로
 2. 큰원반 1개를 A->C로
 3. 작은원반 n-1개를 B->C로

```python
def HanoiTower(n, A, B, C):
    """
    원반 n개르 A에서 C로 이동시키되
    그 과정에서 B 를 활용
    """
    if n==1: 
        print(n,'원반을',A,'에서',C,'로 옮겼습니다')
    else:
        HanoiTower(n-1,A,C,B)
        print(n,'원반을',A,'에서',C,'로 옮겼습니다')
        HanoiTower(n-1,B,A,C)

HanoiTower(3,'A','B','C')
```

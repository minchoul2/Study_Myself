### 재귀

```python
""" 팩토리얼 // 직접 재귀
    사실 팩토리얼은 재귀로 하는게 더 비효율적이다 """
def factorial(n):
    if n>0:
        return n + factorial(n-1)
    else:
        return 1
    
n = int(input())
factorial(n)
```

```python
"""유클리드 호제법 최대공약수"""
def gcd(x,y):
    if y==0:
        return x
    else:
        return gcd(y, x%y)
    
x,y = map(int,input().split(' '))
gcd(x,y)
```

## 8퀸 문제

```python
# 일단 각 열에 퀸을 1개씩 배치
pos = [0]*8

def put():
    """각 열에 배치한 퀸의 위치를 출력"""
    for i in range(8):
        print(f'{pos[i]:2}', end = '')
    print()
    
def set(i):
    """i열에 퀸을 배치"""
    for j in range(8):
        pos[i] = j
        if i == 1:
            put()
        else:
            set(i+1)

set(0)
```

```python
"""행과 열에 퀸을 1개 배치하는 조합을 재귀적으로 나열"""

pos=[0]*8
flag = [False]*8

def put():
    for i in range(8):
        print(f'{pos[i]:2}',end = '')
    print()
    
def set(i):
    for j in range(8):
        if not flag[j]:
            pos[i] = j
            if i == 7:
                put()
            else:
                flag[j] = True
                set(i+1)
                flag[j] = False
                
set(0)
```

```python
pos = [0]*8
flag_a = [False]*8  # 각 행에 퀸을 배치했는지 체크
flag_b = [False]*15 # 대각선 방향으로 퀸을 체크
flag_c = [False]*15 # 대각선 방향으로 퀸을 체크

def put():
    for i in range(8):
        print(f'{pos[i]:2}',end='')
    print()

def put2():
    for i in range(8):
        for j in range(8):
            print('■' if pos[j] == i else '□',end='')
        print()
    print()
    
def set(i):
    for j in range(8):
        if(   not flag_a[j]
          and not flag_b[i+j]
          and not flag_c[i-j+7]):
            pos[i] = j
            if i == 7:
                put2()
                print()
            else:
                flag_a[j] = flag_b[i+j] = flag_c[i-j+7] = True
                set(i+1)
                flag_a[j] = flag_b[i+j] = flag_c[i-j+7] = False

set(0)
                
    
```

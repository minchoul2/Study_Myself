# 이진트리 이진검색트리

## 이진검색트리의 노드

```python
class Node:
    def __init__(self, key, value, left, right):
        self.key = key
        self.value = value
        self.left = left
        self.right = right
        
```

## 이진 검색트리

```python
class BinarySearchTree:
    def __init__(self):
        self.root = None # 루트
    
    def search(self,key):
        p = self.root # 주목노드 p
        
        while True:
            if p is None:
                return None
            if key == p.key:
                return p.value
            elif key < p.key:
                p = p.left
            else:
                p = p.right
    
    # 노드추가함수 : 재귀적인 특석을 살림
    def add(self, key, value): 
        """ node를 루트로 하는 서브트리에 키가 key, 값이 value인 노드 삽입"""
        def add_node(node, key, value):
            if node.left is None:
                return False
            elif key < node.key:
                if node.left is None:
                    node.left = Node(key, value, None, None)
                else:
                    add_node(node.left, key, value)
            else:
                if node.right is None:
                    node.right = Node(key, value, None, None)
                else:
                    add_node(node.rightm key, value)
                    
            return True
            
        if self.root is None:
            self.root = Node(key, value, None, None)
        else:
            return add_node(self.root, key, value)
        
    # 제거함수
    def remove(self,key):
        p = self.root # 주목노드 p; 스캔중인 노드
        parent = None # 스캔중인 노드 p의 부모 노드 # 스캔중인 노드 p의 부모 노드
        is_left_child = True
        
        # 1. 검색 ; 검색 성공시 p는 검색한 노드 참조 & parent는 검색한 노드의 부모 노드 참조
        while True:
            if p is None: # 검색실패
                return False
            
            if key == p.key: # 검색성공
                break
            else:
                parent = p # 내려가기전에 부모 설정
                if key < p.key: # key 쪽이 작으면 
                    is_left_child = True
                    p = p.left 
                else: # key쪽이 크면
                    is_left_child = False
                    p = p.right
                    
        # 2. 자식노드 0 or 1개 일때 삭제            
        if p.left is None: # 왼쪽자식이 없는 노드 삭제 : 자식노드 = 0or1; 주목의 부모노드 - 주목의 오른쪽 자식노드 연결
            if p is self.root:
                self.root = p.right #주목노드가 루트면 일단 right로 가자
            # 원래는 부모-주목-주목오른쪽자식 -> 부모-주목오른쪽자식
            elif is_left_child:
                parent.left = p.right
            else:
                parent.right = p.right
        
        elif p.right is None: # 오른쪽자식이 없는 노드 삭제 : 자식노드 = 0or1; 주목의 부모노드 - 주목의 왼쪽 자식노드 연결
            if p is self.root:
                self.root = p.left
            # 원래는 부모-주목-주목오른쪽자식 -> 부모-주목오른쪽자식
            elif is_left_child:
                parent.left = p.left 
            else:
                parent.right = p.left
        
        # 2. 자식노드 2개 일 떄 삭제    
        else:
            parent = p    
            left = p.left # 서브트리 안에 가장 큰 노드; 일단 p.left로 초기화
            is_left_child = True
            while left.right is not None: # 가장 큰 노드 left를 검색 ; 오른쪽이 큰 값이기 때문에 .right없으면 가장 큰값
                parent = left
                left = left.right
                is_left_child = False # 이렇게 해야지 3.에서 엄한데로 안감
            
            p.key = left.key  # 2. 주목노드 p를 left(p의 왼쪽 서브노드중 가장 큰 값) 로 옮기기
            p.value = left.value
            if is_left_child: # 3. left를 삭제 (parent(처음 left의 parent)와 left가 가지고 있던 자식과 연결)
                parent.left = left.right 
            else:
                parent.right = left.left 
        return True
    
    # 키를 오름차순으로 출력 : 재귀 ; 내림차순 시 reverse = True
    def dump(self, reverse = False):
        
        def print_subtree(node):
            if node is not None:
                print_subtree(node.left)
                print(f'{node.key}:{node.value}')
                print_subtree(node.right)
        
        def print_subtree_rev(node):
            if node is not None:
                print_subtree_rev(node.right)
                print(f'{node.key}:{node.value}')
                print_subtree_rev(node.left)
        
        print_subtree_rev(self.root) if reverse else print_subtree(self.node) # revers인자 값에 따라 다른 함수 호출
        
    # 최소키 반환
    def min_key(self):
        if self.root is None:
            return None
        p = self.root
        while p.left is not None:
            p = p.left
        return p.key
    
    # 최대키 반환
    def max_key(self):
        if self.root is None:
            return None
        p = self.root
        while p.right is not None:
            p = p.right
        return p.key
    
    
```
